```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
#Importando URL

url='https://raw.githubusercontent.com/marcofavero3/dfKaggle-Salary_Data/main/Salary_Data.csv'

df = pd.read_csv(url)

df.head()
```

```python
# Renomeando as colunas
df.rename(columns={
    'YearsExperience': 'XP',
    'Salary': 'Renda',
}, inplace=True)
```

```python
#Verificar dimensão

df.shape
```

```python
#Campos nulos
df.isnull().sum()
```

```python
#Grafico dos Nulos

sns.heatmap(df.isnull() ,cbar=False);
```

```python
#Descrição dos quartis, média e máximo
df.describe()
```

```python
#Plotagem do gráfico de renda
plt.figure(figsize=(10,5))
plt.title('Analise da Renda')
sns.kdeplot( df['Renda'], color='green' );
```

```python
#Plotagem do gráfico de XP
plt.figure(figsize=(10,5))
plt.title('Analise de XP')
sns.kdeplot(df['XP'], color='red');
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.boxplot( df['Renda'] );
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.boxplot( df['XP'] );
```

```python
# Distribuição
plt.figure(figsize=(10, 5))
plt.scatter(data=df, x='Renda', y='XP')
```

```python
plt.figure( figsize=(10, 5) )
sns.regplot( data=df, x='Renda', y='XP');
```

```python
#CORRELAÇÃO

Correlacao = df.corr('pearson')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```

```python
#CORRELAÇÃO

# Correlação
Correlacao = df.corr('spearman')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```
